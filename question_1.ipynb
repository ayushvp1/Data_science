```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, r2_score

data = pd.read_csv("employee performance.csv")
df = pd.DataFrame(data)
x = df[["Age","Years_at_Company","Department","Training_Hours_Last_Year"]]
y = df["Performance_Rating"]
# le = LabelEncoder()
# df["Department"] = le.fit_transform(df["Department"])
ohe = OneHotEncoder(handle_unknown='ignore')
ct = ColumnTransformer(
    transformers=[("encoder", ohe,["Department"])],
)
x = ct.fit_transform(x)
print(df.head())


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=44)
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

acc = accuracy_score(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print("-----------Report----------\n")
print(f"Accuracy score: {acc:.2f}")
print(f"R2 Score: {r2:.4f}")

age = int(input("Enter the age: "))
exp = int(input("Enter the experience: "))
dept = input("Enter the Department: ")
hrs = int(input("Enter the training hours: "))

user_df = pd.DataFrame([["age","exp","dept","hrs"]], columns=["Age","Years_at_Company","Department","Training_Hours_Last_Year"])
# user_df["Department"] = le.fit_transform(user_df["Department"])
user_x = ct.transform(user_df)
predicted_performance = model.predict(user_x)
print(f"\nPredicted Performance: ${predicted_performance}")


```
