```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100)
y = np.tan(x)
plt.plot(x,y, color = 'cyan', linestyle = 'solid', marker='o')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.random.rand(5)
y = np.random.rand(5)
plt.scatter(x,y, color = 'blue', marker ='x')
plt.title("Scatter Plot")
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [3,7,5]
plt.bar(categories, values, color=['red','green','blue'])
plt.title("Bar Plot")
plt.show()
plt.barh(categories,values,color='purple')
plt.show
```

```python
data = np.random.randn(1000)
plt.hist(data,bins=30,color='gold',edgecolor='black')
plt.title("Histogram")
plt.show()
```

```python
sizes = [20,30,25,25]
labels = ['A','B','C','D']
plt.pie(sizes,labels=labels, autopct='%1.1f%%',startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
#stacked area plot

days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
plt.stackplot(days,sleeping,eating,working,playing, labels=['sleep','eat','work','play'])
plt.legend(loc = 'upper left')
plt.title("Stacked Area Plot")
plt.show()
```

```python
#box plot

data = np.random.normal(100,20,200)
plt.boxplot(data)
plt.title("Box plot")
plt.show()
```
